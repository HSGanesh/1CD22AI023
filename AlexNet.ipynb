```python
# -*- coding: utf-8 -*-
"""
Modified AlexNet Architecture
Author: Ganesh
"""

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import (
    Conv2D, MaxPooling2D, Flatten,
    Dense, Dropout, BatchNormalization
)

class ModifiedAlexNet(Sequential):
    def __init__(self, input_shape, num_classes):
        super().__init__()

        # 1️⃣ Modified First Convolution Layer
        self.add(Conv2D(
            96, kernel_size=(7, 7), strides=2,
            padding='same', activation='relu',
            input_shape=input_shape
        ))
        self.add(BatchNormalization())
        self.add(MaxPooling2D(pool_size=(3, 3), strides=2))

        # 2️⃣ Second Convolution Layer
        self.add(Conv2D(256, kernel_size=(5, 5), padding='same', activation='relu'))
        self.add(BatchNormalization())
        self.add(MaxPooling2D(pool_size=(3, 3), strides=2))

        # 3️⃣ Third, Fourth, Fifth Conv Layers
        self.add(Conv2D(384, kernel_size=(3, 3), padding='same', activation='relu'))
        self.add(Conv2D(384, kernel_size=(3, 3), padding='same', activation='relu'))
        self.add(Conv2D(256, kernel_size=(3, 3), padding='same', activation='relu'))
        self.add(MaxPooling2D(pool_size=(3, 3), strides=2))

        # 4️⃣ Flatten
        self.add(Flatten())

        # 5️⃣ Reduced Fully Connected Layers
        self.add(Dense(1024, activation='relu'))
        self.add(Dropout(0.6))
        self.add(Dense(1024, activation='relu'))
        self.add(Dropout(0.6))

        # 6️⃣ Output Layer
        self.add(Dense(num_classes, activation='softmax'))

# Example usage
input_shape = (224, 224, 3)
num_classes = 10  # CIFAR-10 / custom dataset
model = ModifiedAlexNet(input_shape, num_classes)
model.summary()
```
